```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

# Read the image in greyscale mode (let flag = 0)
image_path = 'CT.png'
image_input = cv2.imread(image_path,0)

# Show the input image
plt.imshow(image_input) 

# Define the kernel
kernel_x = np.array([[1,0,-1],
                    [2,0,-2],
                    [1,0,-1]])
kernel_y = np.array([[1,2,1],
                    [0,0,0],
                    [-1,-2,-1]])
dimension_kernel = 3

# Get the information of this image
px = image_input[426,569]
print(px)
print(image_input[426][569])
print(image_input.shape)
row_max = image_input.shape[0]
volumn_max = image_input.shape[1]
data_type = image_input.dtype
print(image_input)
```

```python
def apply_sobel(image, kernel, r_max, v_max, dimension_kernel, data_type):
    center_position = int(dimension_kernel/2)
    image_filtered = np.copy(image)
    # image area loop
    for r in range (0+center_position, r_max-center_position):
        for v in range (0+center_position, v_max-center_position):
            # kernel area loop
            array_tmp1 = np.ones(dimension_kernel,dtype = data_type)
            for w in range (0,dimension_kernel): # row
                for u in range (0,dimension_kernel): # column
                    array_tmp1[u] = image[r-center_position+w][v-center_position+u]
                if w == 0:
                    array_tmp2 = np.copy(array_tmp1)
                else:
                    array_tmp2 = np.vstack((array_tmp2,array_tmp1))
            array_tmp0 = array_tmp2 * kernel
            center_value = np.sum(array_tmp0)
            if center_value <= 0:
                center_value = abs(center_value)
            if center_value >= 255:
                center_value = 255
            image_filtered[r][v] = center_value
    return image_filtered
```

# Remark：
### To be honest, I don't know why we should use the abs value when center_value <= 0, and use 255 when center_value >= 255. I have tried many methods, including asarray(dtype = uint8), not do anything, use 0 when center_value <= 0, etc. They all generate different results. 
### But luckily, when appling above method, it seems that the output image is at least 99% simillar to the sample output. (After observing the array output by cv2.sobel, I discovered that we should use the abs value. As for 255, after experimenting many times, I found out this method eventually.)

```python
def cal_gradient_magnitude(image_x, image_y, row_max, volumn_max):
    image_tmp = np.copy(image_x)
    for r in range (0,row_max):
        for v in range (0,volumn_max):
            value_tmp = math.sqrt(math.pow(image_x[r][v],2) + math.pow(image_y[r][v],2))
            if value_tmp <= 0:
                value = 0
            elif value_tmp >= 255:
                value = 255
            else:
                value = value_tmp
            image_tmp[r][v] = value
    return image_tmp
```

```python
image_filtered_x = apply_sobel(image_input, kernel_x, row_max, volumn_max, dimension_kernel, data_type)
image_filtered_y = apply_sobel(image_input, kernel_y, row_max, volumn_max, dimension_kernel, data_type)
```
